```python
from tkinter import *
import pickle
from tkinter import messagebox
from sklearn.ensemble import AdaBoostClassifier
import numpy as np


root=Tk()
root.title("Diabetes")

l1=Label(root,text="Pregnancy")
l1.grid(row=0,column=0)

var1=IntVar()
text1=Entry(root,textvariable=var1)
text1.grid(row=0,column=1)


l2=Label(root,text="Glucose")
l2.grid(row=1,column=0)

var2=IntVar()
text2=Entry(root,textvariable=var2)
text2.grid(row=1,column=1)

l3=Label(root,text="BP")
l3.grid(row=2,column=0)

var3=IntVar()
text3=Entry(root,textvariable=var3)
text3.grid(row=2,column=1)

l4=Label(root,text="Skin Thickness")
l4.grid(row=3,column=0)

var4=IntVar()
text4=Entry(root,textvariable=var4)
text4.grid(row=3,column=1)

l5=Label(root,text="Insulin")
l5.grid(row=4,column=0)

var5=IntVar()
text5=Entry(root,textvariable=var5)
text5.grid(row=4,column=1)

l6=Label(root,text="BMI")
l6.grid(row=5,column=0)

var6=DoubleVar()
text6=Entry(root,textvariable=var6)
text6.grid(row=5,column=1)


l7=Label(root,text="Diabetes Pedigree Function")
l7.grid(row=6,column=0)

var7=DoubleVar()
text7=Entry(root,textvariable=var7)
text7.grid(row=6,column=1)

l8=Label(root,text="Age")
l8.grid(row=7,column=0)

var8=IntVar()
text8=Entry(root,textvariable=var8)
text8.grid(row=7,column=1)

l9=Label(root,text="Prediction")
l9.grid(row=8,column=0)

text9=Entry(root)
text9.config(state=DISABLED)
text9.grid(row=8,column=1)

def predic():
    filename="finalmodel.sav"
    model = pickle.load(open(filename, 'rb'))
    a=[]
    try:
        a.append(var1.get())
        if var1.get()<0 or var1.get()>15:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in pregnancy cell")
    
    try:
        a.append(var2.get())
        if var2.get()<0 or var2.get()>200:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in Glucose cell")
    
    try:
        a.append(var3.get())
        if var3.get()<0 or var3.get()>125:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in BloodPressure cell")
    
    try:
        a.append(var4.get())
        if var4.get()<0 or var4.get()>100:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in SkinThickness cell")
    
    try:
        a.append(var5.get())
        if var5.get()<0 or var5.get()>850:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in Insulin cell")
    
    try:
        a.append(var6.get())
        if var6.get()<0 or var6.get()>68:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in BMI cell")
    
    try:
        a.append(var7.get())
        if var7.get()<0 or var7.get()>2.5:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in DiabetesPedigreeFunction cell")
    
    try:
        a.append(var8.get())
        if var8.get()<0 or var8.get()>150:
            raise ValueError('Wrong Value')
    except:
        messagebox.showerror("Value Error", "Wrong Value in Age cell")
    predict=model.predict(np.asarray(a).reshape(1, -1))
    text9.configure(state='normal')
    text9.delete(0,END)
    text9.insert(0, predict[0])
    text9.configure(state='disabled')


button=Button(root,text="Predict",command=predic)
button.grid(row=8,column=2)
root.mainloop()
```
